# Linear Regression Physical exercise in working life

```python
#Train Test Split
X = df[['EJERCICIO FISICO']]
y = df['TIEMPO_VIDA_LABORAL']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# Training the Model
from sklearn.linear_model import LinearRegression
# Create an instance of a LinearRegression() model named lm.
lm = LinearRegression()
# Train/fit lm on the training data.
lm.fit(X_train,y_train)
```

```python
# Print out the coefficients of the model
print('Coefficients: \n', lm.coef_)
```

```python
# Predicting Test Data
# Use lm.predict() to predict off the X_test set of the data.
predictions = lm.predict( X_test)
# Create a scatterplot of the real test values versus the predicted values. 
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
# Residuals
# Plot a histogram of the residuals and make sure it looks normally distributed. 
sns.distplot((y_test-predictions),bins=50);
```

```python
# Evaluating the Model
from sklearn import metrics

# Calculate the Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error. 
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
# Recreate the dataframe below. 
coeffecients = pd.DataFrame(lm.coef_,X.columns)
coeffecients.columns = ['Coeffecient']
coeffecients
```

```python
# Interpreting the coefficients:
# Holding all other features fixed, a 1 unit increase in EJERCICIO FISICO is associated with an increase of 0.07896 TIEMPO_VIDA_LABORAL
```

```python
#Kmeans cluster creation
from sklearn.cluster import KMeans
#Define number of clusters
kmeans = KMeans(n_clusters=2)
df.rename(columns={'BURNOUT':'TBURNOUT'}, inplace=True)
#Fit all the model less TBURNOUT column
kmeans.fit(df.drop('TBURNOUT',axis=1))
```

```python
#Guessing cluster center vectors
kmeans.cluster_centers_
```

```python
#Evaluation
#Create a new column for df called Cluster which value 1 for BURNOUT and 0 for not BURNOUT
def converter(cluster):
    if cluster=='Yes':
        return 1
    else:
        return 0
df['Cluster'] = df['TBURNOUT'].apply(converter)
df
```

```python
#Create a confusion matrix and classification report
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(df['Cluster'],kmeans.labels_))
print(classification_report(df['Cluster'],kmeans.labels_))
```
